## Libraries

```python
import pandas as pd # library for data manipulation
import matplotlib.pyplot as plt # library for visualization
import seaborn as sns # library for visualization
sns.set() # this command sets the seaborn chart style as the default

from matplotlib.ticker import PercentFormatter #converts values into percentage format
```

## Bar Chart

```python
df_used_cars = pd.read_csv("bar_chart_data.csv") #if file location isn't the same, specify file locations with /
```

```python
df_used_cars
```

```python
plt.figure(figsize= (9,6)) #spread x axis so labels don't overlap, default figsize is (6.4, 4.8)
plt.bar(x = df_used_cars["Brand"],
       height = df_used_cars["Cars Listings"],
       color = "midnightblue") #rgbwymc
plt.xticks(rotation = 45, fontsize = 13) #rotate x axis labels
plt.yticks(fontsize = 13)
plt.title("Cars Listings by Brand", fontsize = 16, fontweight = "bold") #specify title
plt.ylabel("Number of Listings", fontsize = 13)
plt.show()
plt.savefig("Used Cars Bar.png") #save chart for exporting
```

## Pie Chart

```python
df_fuel_engine_type = pd.read_csv("pie_chart_data.csv")
```

```python
df_fuel_engine_type
```

```python
sns.set_palette('colorblind')
```

```python
plt.figure(figsize = (10,8))
plt.pie(df_fuel_engine_type['Number of Cars'],
       labels = df_fuel_engine_type['Engine Fuel Type'].values,
       autopct = '%.2f%%',
       textprops = {'size' : 'x-large',
                   'fontweight' : 'bold',
                   'rotation' : 30,
                   'color' : 'w'})
plt.legend()
plt.title('Cars by Engine Fuel Type', fontsize = 18, fontweight= "bold")
plt.show()
```

## Stacked Area Chart

```python
df_fuel_engine_types = pd.read_csv("stacked_area_chart_data.csv")
```

```python
df_fuel_engine_types
```

```python
colors = ["#011638", "#7e2987", "#ef2026"]
labels = ["Diesel", "Petrol", "Gas"]
sns.set_style("white") #background colour
plt.figure(figsize = (12,6))
plt.stackplot(df_fuel_engine_types["Year"],
             df_fuel_engine_types["Diesel"],
             df_fuel_engine_types["Petrol"],
             df_fuel_engine_types["Gas"],
             colors = colors,
             edgecolor = "none")
plt.xticks(df_fuel_engine_types["Year"], rotation = 45)
plt.legend(labels = labels, loc = "upper left")
plt.ylabel("Number of Cars", fontsize = 13)
plt.title("Popularity of Engine Fuel Types (1982 - 2016)", fontsize = 14, weight = "bold")
sns.despine() #remove top and right borders
plt.show()
```

## Line Chart

```python
df_spx_ftse_00_10 = pd.read_csv("line_chart_data.csv")
```

```python
df_spx_ftse_00_10
```

```python
# 'Date' is not in datetime format, instead is regarded as a string

df_spx_ftse_00_10["new_date"] = pd.to_datetime(df_spx_ftse_00_10["Date"])
```

```python
df_spx_ftse_00_10["new_date"]
```

```python
labels = ["S&P500", "FTSE 100"]
plt.figure(figsize = (20,8))
plt.plot(df_spx_ftse_00_10["new_date"], df_spx_ftse_00_10["GSPC500"])
plt.plot(df_spx_ftse_00_10["new_date"], df_spx_ftse_00_10["FTSE100"])
plt.title("S&P vs FTSE Returns (2000 - 2010)", fontsize = 14, fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

```python
#choose Q3 and Q4 of 2008 -> slice data

df_spx_ftse_H2_08 = df_spx_ftse_00_10[(df_spx_ftse_00_10.new_date >= '2008-07-01') &
                                     (df_spx_ftse_00_10.new_date <= '2008-12-31')]
```

```python
df_spx_ftse_H2_08
```

```python
labels = ["S&P500", "FTSE 100"]
plt.figure(figsize = (20,8))
plt.plot(df_spx_ftse_H2_08["new_date"], df_spx_ftse_H2_08["GSPC500"], color = "midnightblue")
plt.plot(df_spx_ftse_H2_08["new_date"], df_spx_ftse_H2_08["FTSE100"], color = "crimson")
plt.title("S&P vs FTSE Returns (H2 2008)", fontsize = 14, fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

## Scatter Plot

```python
df_real_estate = pd.read_csv("scatter_data.csv")
```

```python
df_real_estate
```

```python
plt.figure(figsize = (12,8))
scatter = plt.scatter(df_real_estate["Area (ft.)"],
           df_real_estate["Price"],
           alpha = 0.6, #set transparency where a lot of data points overlaps
           c = df_real_estate['Building Type'], #set different colours based on
           cmap = 'viridis') #color mapping
plt.legend(*scatter.legend_elements(), #* pointer
          loc = "upper left",
          title = "Building Type") 
plt.title("Relationship between Area and Price of California Real Estate",
         fontsize = 14,
         weight = "bold")
plt.xlabel("Area (sq. ft.)", weight = "bold")
plt.ylabel("Price (000's of $)")
plt.show()
```

```python
#using seaborn.scatterplot() instead of matplotlib.scatter()

plt.figure(figsize = (12,8))
sns.scatterplot(df_real_estate['Price'],
                df_real_estate['Area (ft.)'],
                hue = df_real_estate['Building Type'], #hue for seaborn
                palette = ['black', 'darkblue', 'purple', 'pink', 'white'],
                s = 100) #size of points
plt.title("Relationship between Area and Price of California Real Estate",
         fontsize = 14,
         weight = "bold")
plt.xlabel("Area (sq. ft.)", weight = "bold")
plt.ylabel("Price (000's of $)")
plt.show()
```

## Regression Plot

```python
df_ad_expenditure = pd.read_csv("scatter_plot_ii.csv")
```

```python
df_ad_expenditure
```

```python
#sns.regplot() - regression plot
#plt.figure(figsize = (10,8))
sns.set(rc = {'figure.figsize' : (9,6)})
sns.regplot(x = "Budget",
           y = "Sales",
           data = df_ad_expenditure,
           scatter_kws = {'color' : 'k'},
           line_kws = {'color' : 'red'})
plt.xlabel("Ad Expenditure in (000's of $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 14, weight = "bold")
plt.show() #grey area around regression line is the 95% confidence interval added by default
```

```python
#sns.lmplot() - linear model plot
# can only support x and y in the form of strings, also known as 'long form' or 'tidy' data
sns.lmplot(x = "Budget",
          y = "Sales",
          data = df_ad_expenditure,
          height = 10, #size -> height
           scatter_kws = {'color' : 'k'},
           line_kws = {'color' : 'red'})
plt.xlabel("Ad Expenditure in (000's of $)")
plt.ylabel("Sales in (000's Units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 14, weight = "bold")
plt.show()
```

## Bar and Line Chart

```python
from matplotlib.ticker import PercentFormatter
```

```python
df_kdnuggets = pd.read_csv("bar_line_chart_data.csv")
```

```python
df_kdnuggets
```

```python
fig, ax = plt.subplots(figsize = (10,7))

ax.bar(df_kdnuggets["Year"],
       df_kdnuggets["Participants"],
       color = "k")
ax.set_ylabel("Number of Participants",
            weight = "bold")
ax.tick_params(axis = "y",
              width = 2,
              labelsize = "large")

ax1 = ax.twinx() #reflect
ax1.set_ylim(0,1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
ax1.plot(df_kdnuggets["Year"],
        df_kdnuggets["Python Users"],
        color = "#b60000",
        marker = "D") #marker shape
ax1.set_ylabel("Python Users",
              color = "#b60000",
              weight = "bold")
ax1.tick_params(axis = "y",
              colors = "#b60000",
              width = 2,
              labelsize = "large")
ax.set_title("KD Nuggets Survey Python Users (2012 - 2019)", fontsize = 14, weight = "bold")
```
